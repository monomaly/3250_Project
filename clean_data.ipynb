```python
#import pandas library as pd
import pandas as pd
```

```python
#Load the dataset
df = pd.read_csv('Tech_Use_Stress_Wellness.csv')
```

```python
#display the current basic info of rows and columns
df.info()
```

```python
#show current shape of dataset to compare before and after it is cleaned
print(f"Original shape: {df.shape}")
```

```python

#check for missing values in each coloumns
print("Missing values per column:")
print(df.isnull().sum())
```

```python
#define the columns relevant to our interests
columns_to_keep = [
    'uses_wellness_apps', 'mindfulness_minutes_per_day', 'sleep_quality','stress_level', 'sleep_duration_hours', 'caffeine_intake_mg_per_day',
    'gaming_hours', 'entertainment_hours', 'phone_usage_hours', 'laptop_usage_hours', 'gender', 'daily_screen_time_hours', 'mood_rating',
    'physical_activity_hours_per_week', 'mental_health_score', 'weekly_depression_score', 'social_media_hours', 'work_related_hours'
]
```

```python

#filter the dataset and remove any rows with empty values (there turned out to be no null values)
cleaned_df = df[columns_to_keep].dropna()
```

```python
#save the cleaned dataset to a new file
cleaned_df.to_csv('Cleaned_Tech_Use_Stress_Wellness.csv', index=False)
```

```python
#show the cleaned dataset shape and its info to confirm updates
print("\nDataset cleaned successfully!")
print(f"Final shape: {cleaned_df.shape}")
cleaned_df.info()
```
